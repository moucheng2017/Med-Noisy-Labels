## Convert imgs from .jpg to .tif and populate images folder

```python
img_size = (256, 192)
```

```python
path_old = '/Users/gathanasiou/data/eurova/cumulus_database/oocytes/'
path_new = '/Users/gathanasiou/git/Learn_Noisy_Labels_Medical_Images/oocytes_gent/train/images/'
```

```python
path_masks = '/Users/gathanasiou/data/eurova/cumulus_database/masks/AR/immature/'
list_names = []
for img in os.listdir(path_masks):
    if img.endswith('.png'):
        print(img)
        list_names.append(str(img))
        
print(len(list_names))
```

```python
# errors in immature only, for every 1.*
# for mature it works fine for all :S :S :S
#list_names.remove('1.17.png')
```

```python
phase = 'mature/'
for img in list_names:
    if img.endswith('.png'):
        
        print(img)
        im = Image.open(os.path.join(path_old + phase, img))
        print(im.size)
        tif_img_path = path_new + str(img[:-4]) + '.tif'
        print(tif_img_path)
        
        im.thumbnail(img_size)
        im.save(tif_img_path)
        
        
```

```python
path_AR = '/Users/gathanasiou/data/eurova/cumulus_database/masks/AR/mature/'
path_HS = '/Users/gathanasiou/data/eurova/cumulus_database/masks/HS/mature/'
path_SG = '/Users/gathanasiou/data/eurova/cumulus_database/masks/SG/mature/'
path_output_AR = '/Users/gathanasiou/git/Learn_Noisy_Labels_Medical_Images/oocytes_gent/train/AR/'
path_output_HS = '/Users/gathanasiou/git/Learn_Noisy_Labels_Medical_Images/oocytes_gent/train/HS/'
path_output_SG = '/Users/gathanasiou/git/Learn_Noisy_Labels_Medical_Images/oocytes_gent/train/SG/'
```

```python
for img in list_names:
    if img.endswith('.png'):
        
        #AR
        im = Image.open(os.path.join(path_AR, img))
        im.thumbnail(img_size)
        im.save(path_output_AR + img[:-4] + '.tif')
        #HS
        im = Image.open(os.path.join(path_HS, img))
        im.thumbnail(img_size)
        im.save(path_output_HS + img[:-4] + '.tif')
        #SG
        im = Image.open(os.path.join(path_SG, img))
        im.thumbnail(img_size)
        im.save(path_output_SG + img[:-4] + '.tif')
```

```python
path_avrg = '/Users/gathanasiou/data/eurova/cumulus_database/masks/avrg/mature/'
path_output_avrg = '/Users/gathanasiou/git/Learn_Noisy_Labels_Medical_Images/oocytes_gent/train/avrg/'
```

```python
for img in list_names:
    if img.endswith('.png'):
        
        #avrg
        im = Image.open(os.path.join(path_avrg, img)).convert('L')
        im.thumbnail(img_size)
        im.save(path_output_avrg + img[:-4] + '.tif')
```

```python
import random

test_list = []
validate_list = []

for i in range(0, 5):
    test_list.append(random.choice(list_names))
    validate_list.append(random.choice(list_names))
    
print(test_list)
print(validate_list)
```
